```python
from pandas import Series, DataFrame
import pandas as pd
```

```python
# Считывание таблицы movies
movies = pd.read_table('C:\\Users\\tosha\\movielens\\movies.dat', sep='::', names=['MovieID', 'Title', 'Genre'], engine='python')
movies
```

```python
# Считывание таблицы ratings
ratings = pd.read_table('C:\\Users\\tosha\\movielens\\ratings.dat', sep='::', names=['UserID', 'MovieID', 'Rating', 'Timestap'], engine='python')
ratings
```

```python
# Считывание таблицы users
users = pd.read_table('C:\\Users\\tosha\\movielens\\users.dat', sep='::', names=['UserID', 'Gender', 'Age', 'Occepation', 'Postcode'], engine='python')
users
```

```python
# Объединение трёх таблиц в одну
data = pd.merge(ratings, users)
data = pd.merge(data, movies)
data
```

```python
# Средний рейтинг среди мужчин и женщин
mean_ratings_gender = pd.pivot_table(data, values='Rating', index='Title', columns='Gender', aggfunc='mean', fill_value=0)
mean_ratings_gender
```

```python
# Средний рейтинг среди людей различного возраста
mean_ratings_age = pd.pivot_table(data, values='Rating', index='Title', columns='Age', aggfunc='mean', fill_value=0)
mean_ratings_age
```

```python
# Рейтинг среди женщин, отсортированный по убыванию
sort_ratings_F = mean_ratings_gender.sort_values(by='F', ascending=False)
sort_ratings_F
```

```python
# Рейтинг среди 56-ти летних людей, отсортированный по убыванию
sort_ratings_age = mean_ratings_age.sort_values(by=56, ascending=False)
sort_ratings_age
```

```python
# Список фильмов, кооторые получили не менее 250 оценок
table_movies = data.groupby('Title').agg({'Rating': 'count'})
table_movies = table_movies[table_movies >= 250].dropna(axis=0)
table_movies
```

```python
# Список фильмов, которые получилт высшую оценку среди женщин и низшую среди мужчин
mean_ratings_gender['difference'] = mean_ratings_gender['F'] - mean_ratings_gender['M']
mean_ratings_gender.sort_values(by='difference', ascending=False).head(15)
```

```python
# Список фильмов, которые получилт высшую оценку среди мужчин и низшую среди женщин
mean_ratings_gender = mean_ratings_gender.drop(labels='difference', axis=1)
mean_ratings_gender['difference'] = mean_ratings_gender['M'] - mean_ratings_gender['F']
mean_ratings_gender.sort_values(by='difference', ascending=False).head(15)
```
