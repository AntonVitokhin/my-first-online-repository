```python
import random
```

```python
class Bank():
    'Класс-симулятор манипуляций с картой в отделе банка'
    
    all = []
    
    def __init__(self, name, age, money_balance):
        assert money_balance >= 0, 'Стартовый капитал не может быть отрицательным'
        assert age > 13, 'Ваш возраст не позволяет стать клиентом банка'
        
        self.name = name
        self.age = age
        self.money_balance = money_balance
        self.all.append(self)
        
    def __del__(self):
        self.all.remove(self)
        print(f'{self.name} deleted')
        
        
    def __repr__(self):
        if getattr(self, 'card', False) == 0:
            return f'Данные о клиенте {self.name}. Возраст: {self.age}, денежные средства: {self.money_balance}. У вас нет карты'
        else:
            return f'Данные о клиенте {self.name}. Возраст: {self.age}, денежные средства: {self.money_balance}. Номер карты: {self.card}, cvv-код: {self.card_cvv}, баланс карты: {self.card_balance}'
        
        
    def get_card(self, card_password):  # Оформить карту
        assert len(card_password) == 4, 'Пароль должен состоять из 4 цифр'
        assert card_password.isdigit(), 'Пароль должен состоять только из цифр'
        
        number_card = ''
        
        for i in range(1, 17):
            number_card += str(random.randint(0, 9))
            if i % 4 == 0:
                number_card += ' '
                
        self.a = ['card', 'card_balance', 'card_cvv', 'card_password']        
        self.card = number_card.rstrip()
        self.card_balance = 0
        self.card_cvv = random.randint(100, 999)
        self.card_password = card_password
        print(f'Вы оформили карту {self.card}')
        
        
    def deposit_money(self, money, your_card_password):  # Пополнить баланс карты
        assert money > 0, 'Сумма пополнения должна быть больше нуля'
        
        if self.card_password != your_card_password:
            print('Пароль неверный')
        elif money > self.money_balance:
            print(f'Пополнение карты {self.card} невозможно из-за отсутствия нужного количества средств')
        else:
            self.card_balance += money
            self.money_balance -= money
            print(f'Ваша карта {self.card} успешно пополнена на сумму {money}')
        
        
    def withdraw_money(self, money, your_card_password):  # Снять денгьги с карты
        assert money > 0, 'Сумма снятия должна быть больше нуля'
        
        if self.card_password != your_card_password:
            print('Пароль неверный')
        elif money > self.card_balance:
            print(f'Снятие невозможно из-за отсутствия нужного количества средств на карте {self.card}')
        else:
            self.card_balance -= money
            self.money_balance += money
            print(f'С вашей карты {self.card} были списаны средства на сумму {money}')
        
        
    def close_card(self, your_card_password):  # Закрыть карту
        if self.card_password != your_card_password:
            print('Пароль неверный')
        else:
            print(f'Карта {self.card} была закрыта')
            self.money_balance += self.card_balance
        
            for j in self.a:
                delattr(self, '%s'%j)
        
            delattr(self, 'a')
        
        
    def how_card_balance(self):  # Узнать баланс карты
        return self.card_balance
```

```python
class Mobile_bank(Bank):
    'Класс-симулятор манипуляций с картой в мобильном приложении банка'
    
    
    def __init__(self, name, age, money_balance, password, status=0):
        super().__init__(name, age, money_balance)
        self.password = password
        self.status = status
        
        
    def connect(self, password):  # Подключиться к интернету и войти в приложение с использованием пароля
        if self.password == password:
            self.status = 1
            print(f'Добро пожаловать {self.name}')
        else:
            print('Пароль неверный')
    
    
    def disconnect(self):  # Отключиться от интернета
        self.status = 0
        
        
    def profile(self):  # Посмотреть свой профиль в приложении
        return Bank.all[Bank.all.index(self)]
        
        
    def get_card(self, card_password):  # Оформить карту в приложении
        if self.status == 0:
            print('Ваш статус offline, пожалуйста, подключитесь к интернету')
        else:
            super().get_card(card_password)
    
    
    def deposit_money(self, money, your_card_password):  # Пополнить баланс карты через приложение
        if self.status == 0:
            print('Ваш статус offline, пожалуйста, подключитесь к интернету')
        else:
            super().deposit_money(money, your_card_password)
            
    
    def close_card(self, your_card_password):  # Закрыть карту через приложение
        if self.status == 0:
            print('Ваш статус offline, пожалуйста, подключитесь к интернету')
        else:
            super().close_card(your_card_password)
            
# Метода снятия с карты нет, потому что это не возможно сделать через приложение
```
